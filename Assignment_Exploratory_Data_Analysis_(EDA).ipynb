```python
import pandas as pd
import numpy as np
import seaborn as sns                      
import matplotlib.pyplot as plt   
df = pd.read_csv("zomato.csv",encoding='latin-1') 
df.head() 
```

```python
df.dtypes
```

```python
df = df.rename(columns={"Average Cost for two": "Avg_Cost", "Price range ": "Price", "Aggregate rating": "Rating", "Restaurant Name":"Name" })
df.head(5)
```

```python
pd.unique(df['Rating color'])
```

```python
df = df.drop(['Rating color'], axis=1)
df.head(5)
df = df.drop(['Restaurant ID'], axis=1)
df.head(5)
```

```python

duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
print(df.isnull().sum())
```

```python
sns.boxplot(x=df['Avg_Cost'])
plt.boxplot(df['Avg_Cost'], showfliers=True)
plt.title("Boxplot with Default Whiskers")
plt.show()
```

```python
plt.boxplot(df['Avg_Cost'], whis=3, showfliers=True)
plt.title("Boxplot with Adjusted Whiskers")
plt.show()
```

```python
plt.boxplot(df['Avg_Cost'], showfliers=False)
plt.plot(np.ones_like(df['Avg_Cost']) * 1, df['Avg_Cost'], 'r+', markersize=5)
plt.title("Boxplot with Marked Outliers")
plt.show()
```

```python
from scipy.stats.mstats import winsorize

winsorized_data = winsorize(df['Avg_Cost'], limits=[0.05, 0.05])

plt.boxplot(winsorized_data, showfliers=True)
plt.title("Boxplot of Winsorized Data")
plt.show()
```

```python
df.Name.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Total Number of Restaurant")
plt.ylabel('Number of Restaurant')
plt.xlabel('Name');
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(y=df['Rating'],x=df['Name'])
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,annot=True)
```
